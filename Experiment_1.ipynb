```python
# Define a function to plot histograms and box plots for each feature
def plot_feature_distributions(data):
    # Set the number of features
    features = data.columns.tolist()

    # Create subplots
    num_features = len(features)
    fig, axes = plt.subplots(num_features, 2, figsize=(12, 5 * num_features))

    for i, feature in enumerate(features):
        # Histogram
        sns.histplot(data[feature], ax=axes[i, 0], bins=30, kde=True)
        axes[i, 0].set_title(f'Distribution of {feature}')
        axes[i, 0].set_xlabel(feature)
        axes[i, 0].set_ylabel('Frequency')

        # Box Plot
        sns.boxplot(x=data[feature], ax=axes[i, 1])
        axes[i, 1].set_title(f'Box Plot of {feature}')

    plt.tight_layout()
    plt.show()

# Call the function to plot distributions
plot_feature_distributions(df)
```
