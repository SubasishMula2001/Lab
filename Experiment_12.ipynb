```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/campaign_responses.csv')
```

```python
data.info()
```

```python
data.head(5)
```

```python
# feature engineering
data['income_per_dependent'] = data['annual_income'] / (data['no_of_children'] + 1)
```

```python
# Display the first few rows to verify the new feature
print("\nData with New Feature:")
print(data[['annual_income', 'no_of_children', 'income_per_dependent']].head())
```

```python
# Compute the correlation matrix, considering only numerical features
correlation_matrix = data.corr(numeric_only=True)

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
