```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Function to visualize histograms and box plots for numerical columns
def plot_distribution_and_boxplot(df, columns):
    for col in columns:
        plt.figure(figsize=(12, 4))
        plt.subplot(1, 2, 1)
        sns.histplot(df[col], kde=True, color='skyblue')
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        
        plt.subplot(1, 2, 2)
        sns.boxplot(x=df[col], palette='Set2')
        plt.title(f'Box Plot of {col}')
        plt.xlabel(col)
        
        plt.tight_layout()
        plt.show()

# Step 1: Visualize Distributions and Box Plots Before Outlier Removal
print("Visualizations Before Outlier Removal")
plot_distribution_and_boxplot(cars_df, ['Year', 'Mileage'])

# Step 2: Remove Outliers Using Modified IQR Threshold
def remove_outliers_iqr(data, columns, multiplier=1.5):
    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - multiplier * IQR
        upper_bound = Q3 + multiplier * IQR
        # Remove rows that are outside the bounds
        data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
    return data

# Apply custom IQR threshold (e.g., 2.0 or 3.0) for Year and Mileage
cars_df_no_outliers = remove_outliers_iqr(cars_df.copy(), ['Year', 'Mileage'], multiplier=3.0)

# Step 3: Visualize Distributions and Box Plots After Outlier Removal
print("Visualizations After Outlier Removal")
plot_distribution_and_boxplot(cars_df_no_outliers, ['Year', 'Mileage'])

# Step 4: Save the cleaned dataset with outliers removed
cleaned_file_path = "cleaned_cars_data_no_outliers.csv"
cars_df_no_outliers.to_csv(cleaned_file_path, index=False)
print(f"\nCleaned dataset saved as '{cleaned_file_path}' for further analysis.")
```
